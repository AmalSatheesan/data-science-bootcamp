### Temperature Conversion

```python
def convert_temperature(temp,unit_from,unit_to):
    if unit_from == "C":
        if unit_to == "F":
            return (temp * 9/5) + 32
        elif unit_to == "K":
            return temp + 273.15
    elif unit_from == "F":
        if unit_to == "C":
            return (temp - 32) * 5/9
        elif unit_to == "K":
            return (temp - 32) * 5/9 + 273.15
    elif unit_from == "K":
        if unit_to == "C":
            return temp - 273.15
        elif unit_to == "F":
            return (temp - 273.15) * 9/5 + 32
    return None

convert_temperature(100, "C", "F")  # Example usage
```

### Password Strength Check

```python
def is_strong_password(password):
    """this function checks if the password is strong"""
    if len(password) <8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char in "!@#$%^&*()-+" for char in password):
        return False
    return True
is_strong_password("Password123!")  # Example usage
```

### Total Cost of items in a shopping cart

```python
def calculate_cart_total(cart):
    """ This function calculates the total cost of items in a shopping cart """
    total = 0
    for item in cart:
        total+= item['price'] * item['quantity']
    return total

cart = [
{'name': 'Apple', 'price': 0.5, 'quantity': 4}, 
       {'name': 'Banana', 'price': 0.3, 'quantity': 6},
       {'name': 'Orange', 'price':0.8, 'quantity': 5}
]

calculate_cart_total(cart)  # Example usage
```

### palindrome or not

```python
def is_palindrome(s):
    """' This function checks if a string is a palindrome """
    s= s.lower().replace(" ","")
    return s == s[::-1]
is_palindrome("A man a plan a canal Panama")  # Example usage
```

### Factorial of a num

```python
def factorial(n):
    """This function returns the factorial of a number"""
    if n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)
    
factorial(5)  # Example usage
```

### Functione to read a file and find the num of characters

```python
def count_char_frequency(file_path):
    """This function count frequency of each character in a text file  """
    frequency ={}
    with open(file_path, 'r') as file:
        text = file.read()
        #text = text.lower()
        for char in text:
            if char in frequency:
                frequency[char] += 1
            else:
                frequency[char] = 1 

    return frequency

file_path = 'example.txt'
count_char_frequency('example.txt')  # Example usage   
```

```python
#import string
def count_word_frequency(file_path):
    """This function count frequency of each word in a text file  """
    frequency_1 ={}
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?;"\'()[]{}<>@#$%^&*-_=+`~')
                #word = word.lower().strip(string.punctuation)
                frequency_1[word] = frequency_1.get(word,0) +1
    return frequency_1

file_path = 'example.txt'
count_word_frequency('example.txt')  # Example usage 

```
